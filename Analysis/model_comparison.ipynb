**Fisher z-trasformation**

```python
r_subjects_all = []
r_subjects_noWO = []
r_subjects_nofreq = []
r_subjects_nosurp = []
r_subjects_nodiss = []


r4= mTRF4.r.mean()
r5= mTRF5.r.mean()
r6= mTRF6.r.mean()
r7= mTRF7.r.mean()
r18= mTRF18.r.mean()
r19 = mTRF19.r.mean()

#-----------------------------
r4_noWO= mTRF4_noWO.r.mean()
r5_noWO= mTRF5_noWO.r.mean()
r6_noWO= mTRF6_noWO.r.mean()
r7_noWO= mTRF7_noWO.r.mean()
r18_noWO= mTRF18_noWO.r.mean()
r19_noWO = mTRF19_noWO.r.mean()

#------------------------------
r4_nofreq= mTRF4_nofreq.r.mean()
r5_nofreq= mTRF5_nofreq.r.mean()
r6_nofreq= mTRF6_nofreq.r.mean()
r7_nofreq= mTRF7_nofreq.r.mean()
r18_nofreq= mTRF18_nofreq.r.mean()
r19_nofreq= mTRF19_nofreq.r.mean()

#------------------------------
r4_nosurp= mTRF4_nosurp.r.mean()
r5_nosurp= mTRF5_nosurp.r.mean()
r6_nosurp= mTRF6_nosurp.r.mean()
r7_nosurp= mTRF7_nosurp.r.mean()
r18_nosurp= mTRF18_nosurp.r.mean()
r19_nosurp= mTRF19_nosurp.r.mean()

#------------------------------
r4_nodiss= mTRF4_nodiss.r.mean()
r5_nodiss= mTRF5_nodiss.r.mean()
r6_nodiss= mTRF6_nodiss.r.mean()
r7_nodiss= mTRF7_nodiss.r.mean()
r18_nodiss= mTRF18_nodiss.r.mean()
r19_nodiss= mTRF19_nodiss.r.mean()


r_subjects_noWO.extend([r4_noWO, r5_noWO, r18_noWO, r7_noWO, r6_noWO, r19_noWO])
r_subjects_nofreq.extend([r4_nofreq, r5_nofreq, r18_nofreq, r7_nofreq, r6_nofreq, r19_nofreq])
r_subjects_nosurp.extend([r4_nosurp, r5_nosurp, r18_nosurp, r7_nosurp, r6_nosurp, r19_nosurp])
r_subjects_nodiss.extend([r4_nodiss, r5_nodiss, r18_nodiss, r7_nodiss,  r6_nodiss, r19_nodiss])
r_subjects_all.extend([r4, r5, r18, r7, r6, r19])


# fisher transofrmation
z_all = np.arctanh(r_subjects_all)
z_noWO = np.arctanh(r_subjects_noWO)
z_nofreq = np.arctanh(r_subjects_nofreq)
z_nosurp = np.arctanh(r_subjects_nosurp)
z_nodiss = np.arctanh(r_subjects_nodiss)


order = [r4, r5, r18, r7, r6, r19]
```

**Delta prediction** (difference)

How much worse or better does the model get when I remove predictor *x*?
- Δ ≈ 0 → predictor adds no unique information

- Δ > 0 → predictor adds unique information

- Δ < 0 → predictor hurts prediction

```python
delta_noWO   = z_all - z_noWO
delta_nofreq = z_all - z_nofreq
delta_nosurp = z_all - z_nosurp
delta_nodiss = z_all - z_nodiss
```

**T-test (onse-sample vs 0)**

```python
# Two-tailed one-sample t-tests
print("WO:",   ttest_1samp(delta_noWO,   0))
print("Freq:", ttest_1samp(delta_nofreq, 0))
print("Surp:", ttest_1samp(delta_nosurp, 0))
print("Diss:", ttest_1samp(delta_nodiss, 0))
```
