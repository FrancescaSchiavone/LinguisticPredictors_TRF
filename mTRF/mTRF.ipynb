```python
import numpy as np
from scipy.io import loadmat
import pickle

# Carico dati EEG da Matlab
raw = loadmat(path_EEG19)
x = raw["CutEEG_Zsc"]  # shape: (1, n_trials)

n_trials = x.shape[1]  # es. 15 trial
trials_list = [x[0, i] for i in range(n_trials)]  # ogni trial (6000, 65)
trials_list = [trial.T for trial in trials_list]   # (65, 6000)

# Stack in array 3D: (trials, channels, time)
data_trials = np.stack(trials_list, axis=0, dtype = np.float64) # (15, 65, 6000)
data_trials = data_trials.astype(np.float64)


#Asse temporale
srate = 100  # Hz, come hai detto
n_samples = data_trials.shape[2]
tmin = 0.0
tstep = 1 / srate  # 0.01 s
time = eel.UTS(tmin, tstep, n_samples)

# NDVar EEG
case = eel.Case(n_trials)
n_channels = data_trials.shape[1]
sensor = eel.Sensor.from_montage('GSN-HydroCel-65_1.0')

eeg = eel.NDVar(data_trials, dims=(case, sensor, time), name='EEG')

# Carico predittore 

# --- WO gauss ---
with open(path_WO_gauss19, "rb") as f:
    dataWO = pickle.load(f)

predictor_arrayWO = dataWO["trials"].astype(np.float64)

predictor_ndWO = eel.NDVar(
    predictor_arrayWO,
    dims=(case, time),
    name='Word Onset'
)

# --- Word Frequency ---
with open(path_wordfreq19, "rb") as f:
    dataFreq = pickle.load(f)

predictor_arrayFreq = dataFreq["trials"].astype(np.float64)

predictor_ndFreq = eel.NDVar(
    predictor_arrayFreq,
    dims=(case, time),
    name='Word Frequency'
)

# --- Surprisal ---
with open(path_surp19, "rb") as f:
    dataSurp = pickle.load(f)

predictor_arraySurp = dataSurp["trials"].astype(np.float64)

predictor_ndSurp = eel.NDVar(
    predictor_arraySurp,
    dims=(case, time),
    name='Surprisal'
)


# --- Dissimilarity ---
with open(path_diss19, "rb") as f:
    dataDiss = pickle.load(f)

predictor_arrayDiss = dataDiss["trials"].astype(np.float64)

predictor_ndDiss = eel.NDVar(
    predictor_arrayDiss,
    dims=(case, time),
    name='Semantic dissimilarity'
)

# Cross-validation
trf_cv = eel.boosting(eeg, [predictor_ndWO, predictor_ndFreq, predictor_ndSurp, predictor_ndDiss], tstart=-0.200, tstop= 0.600, partitions=6, test=True, delta=0.003, partition_results=True, basis=0.08, selective_stopping=4, scale_data=True)

out_path = r'C:\Users\schia\OneDrive - Alma Mater Studiorum Università di Bologna\Desktop\mTRF\\Output\\mTRF19.pkl'

eel.save.pickle(trf_cv, out_path)
```

```python
title = f"Mean exp: {trf_cv.proportion_explained.mean('sensor'):.2%}"
p = eel.plot.Topomap(trf_cv.proportion_explained, clip='circle', title=title)
pcb = p.plot_colorbar('Proportion explained')

title = f"Mean r: {trf_cv.r.mean('sensor'):.2f}"
p = eel.plot.Topomap(trf_cv.r, clip='circle', title=title)
pcb = p.plot_colorbar()

mean_r = trf_cv.r.mean()
mean_prop_expl = trf_cv.proportion_explained.mean()
print(mean_r, mean_prop_expl)
```
